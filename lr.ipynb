```python
import pandas as pd
df_raw = pd.read_csv("Daten/winequality-red.csv", delimiter = ";" )
df_raw
```

```python
df = (df_raw - df_raw.mean()) / df_raw.std()
```

```python
df
```

```python
import numpy as np

X = df.drop("quality", axis=1)
X.insert(0, "ones", 1)
X
```

```python
y = df["quality"]
y
```

```python
A = X.T @ X

b = X.T @ y

w = np.linalg.solve(A, -b)
w
```
